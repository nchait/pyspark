```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('tree_ex').getOrCreate()
```

```
from pyspark.ml.classification import RandomForestClassifier, GBTClassifier, DecisionTreeClassifier
from pyspark.ml.feature import VectorAssembler, VectorIndexer, OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.sql.functions import year, month, dayofmonth
```

```
df = spark.read.format("libsvm").load("/FileStore/tables/sample_libsvm_data-2.txt")
df.describe().show() 
```

```
train, test = df.randomSplit([.7,.3])
```

```
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier(numTrees=100)
gbt = GBTClassifier()
```

```
dtcModel = dtc.fit(train)
rfcModel = rfc.fit(train)
gbtModel = gbt.fit(train)
```

```
dtcPreds = dtcModel.transform(test)
rfcPreds = rfcModel.transform(test)
gbtPreds = gbtModel.transform(test)
```

```
dtcPreds.show()
```

```
rfcPreds.show()
```

```
gbtPreds.show()
```

```
acc_eval = MulticlassClassificationEvaluator(metricName='accuracy')
```

```
acc_eval.evaluate(dtcPreds)
```

```
acc_eval.evaluate(rfcPreds)
```
