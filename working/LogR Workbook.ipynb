```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('logr_book').getOrCreate()
```

```
from pyspark.ml.classification import LogisticRegression
```

```
df = spark.read.format("libsvm").load("/FileStore/tables/sample_libsvm_data-1.txt")
df.describe().show()
```

```
logReg = LogisticRegression()
```

```
fitted_LigReg = logReg.fit(df)
```

```
sum = fitted_LigReg.summary
```

```
sum.predictions.printSchema()
```

```
sum.predictions.show()
```

```
train, test = df.randomSplit([.7,.3])
```

```
FlogReg = LogisticRegression()
Ffitted_LogReg = FlogReg.fit(train)
```

```
predsWlabels= Ffitted_LogReg.evaluate(test)
predsWlabels.predictions.show()
```

```
from pyspark.ml.evaluation import (BinaryClassificationEvaluator, MulticlassClassificationEvaluator)
```

```
eval = BinaryClassificationEvaluator()
```

```
roc = eval.evaluate(predsWlabels.predictions)
```
